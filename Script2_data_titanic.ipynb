```python
#importation the librairy of pandas and os
import pandas as pd
import os
```

```python
#loading a csv file
excel_link = './Downloads/titanic.csv'
df= pd.read_csv(excel_link)
```

```python
#showing the data frame 
df
```

```python
#showing a list of columns for data frame
df_columns = df.columns
list(df_columns)
```

```python
#dropping the columns that we don't need for our analysing
columns_drop=['Pclass', 'Name', 'Ticket', 'Fare','Cabin','Embarked']
df=df.drop(columns = columns_drop)
df
```

```python
#combine SibSp and Parch in one columns name's TotalFamily
df['TotalFamily'] = df['SibSp'] + df['Parch']
df
```

```python
#dropping the two columns SibSp and Parch bcs we don't need them
df_drop_fam = ['SibSp','Parch']
df = df.drop(columns=df_drop_fam)
df
```

```python
#dropping the rows in survived columns that have a missing values
df['Survived'].dropna(axis=0, inplace=True)
df #we see that we don't have a missing values in survived columns
```

```python
#converting a female to number 0 and male to number 1
df['Sex'] = df['Sex'].replace({'female': 0, 'male': 1})
df
```

```python
#sauvgarding this data frame in excel file name's titanic01
df.to_excel('titanic01.xlsx', index=False)
```

```python
#impoti a lybrairie matplotlib for visualization the data
import matplotlib.pyplot as plt
# Group the data by the 'Survived' column and count the number of occurrences
survived_counts = df['Survived'].value_counts()
# Create a bar plot to visualize the number of people who survived
plt.bar(survived_counts.index, survived_counts.values)
plt.xlabel('Survived')
plt.ylabel('Number of people')
plt.title('Number of People who Survived')
plt.xticks(survived_counts.index, ['No', 'Yes'])  # Assuming 'Survived' column has values 0 and 1
plt.show()
```

```python
#visualization the sex distribution with pie chart
sex_counts = df['Sex'].value_counts()
sizes = sex_counts  # Sizes of the sectors
labels = ['Female', 'Male']  # Labels for the sectors
colors = ['pink', 'lightblue']  # Colors for the sectors

# Create the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures a circular shape
plt.title('Sex Distribution')

# Display the chart
plt.show()
```

```python
# Calculate the total family size
total_family_size = df['TotalFamily'].sum()

# Plot the total family size
plt.bar('Total Family', total_family_size)

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Family Size')
plt.title('Total Family Size')

# Display the chart
plt.show()
```

```python
# Analyze survival rates for passengers traveling alone vs. with family
alone_survived = df.loc[(df['TotalFamily'] == 0) & (df['Survived'] == 1)]
alone_not_survived = df.loc[(df['TotalFamily'] == 0) & (df['Survived'] == 0)]
with_family_survived = df.loc[(df['TotalFamily'] > 1) & (df['Survived'] == 1)]
with_family_not_survived = df.loc[(df['TotalFamily'] > 1) & (df['Survived'] == 0)]

# Calculate survival rates
alone_survival_rate = len(alone_survived) / len(df.loc[df['TotalFamily'] == 0])
with_family_survival_rate = len(with_family_survived) / len(df.loc[df['TotalFamily'] > 1])

# Plotting the survival rates
labels = ['Alone', 'With Family']
survival_rates = [alone_survival_rate, with_family_survival_rate]

plt.bar(labels, survival_rates)
plt.xlabel('Passenger Type')
plt.ylabel('Survival Rate')
plt.title('Survival Rate: Alone vs. With Family')
```
